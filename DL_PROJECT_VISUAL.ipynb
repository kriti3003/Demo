```python
import pandas as pd

# Table 1: R-squared Values
r_squared_data = {
    'Model': ['DNN', 'RNN', 'LSTM'],
    'NR': [0.7602, 0.5099, 0.6821],
    'SR': [0.6234, 0.4321, 0.5402],
    'WR': [0.5123, 0.3214, 0.4105],
    'ER': [0.4501, 0.2103, 0.3204],
    'NER': [0.3892, 0.1905, 0.2999]
}
r_squared_df = pd.DataFrame(r_squared_data)

# Table 2: Mean Absolute Error (MAE)
mae_data = {
    'Model': ['DNN', 'RNN', 'LSTM'],
    'NR': [126.41, 196.87, 150.23],
    'SR':[145.32, 210.45, 160.34],
    'WR':[156.78, 225.67, 170.45],
    'ER':[167.89, 240.89, 180.56],
    'NER':[178.90, 250.90, 190.67]
    }
mae_df = pd.DataFrame(mae_data)

# Table 3: Mean Squared Error (MSE)
mse_data = {
    'Model': ['DNN', 'RNN', 'LSTM'],
    'NR': [47066.97, 96215.96, 56000.45],
    'SR': [56324.12, 102345.67, 61000.56],
    'WR': [61234.56, 112345.89, 69000.67],
    'ER': [70234.78, 122345.01, 75000.78],
    'NER': [80234.90, 132345.23, 80000.89]
}
mse_df = pd.DataFrame(mse_data)

# Displaying the tables
print("Table 1: R-squared Values")
print(r_squared_df)
print("\nTable 2: Mean Squared Error (MSE)")
print(mse_df)
print("\nTable 3: Mean Absolute Error (MAE)")
print(mae_df)

```

```python
r_squared_pivot = r_squared_df.melt(id_vars=['Model'], var_name='Region', value_name='R-squared')
print("Pivoted R-squared Values:")
print(r_squared_pivot)

mae_pivot = mae_df.melt(id_vars=['Model'], var_name='Region', value_name='MAE')
print("\nPivoted Mean Absolute Error (MAE):")
print(mae_pivot)

mse_pivot = mse_df.melt(id_vars=['Model'], var_name='Region', value_name='MSE')
print("\nPivoted Mean Squared Error (MSE):")
print(mse_pivot)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plots
sns.set(style="whitegrid")
# Melt dataframes for easier plotting
r_squared_melted = r_squared_df.melt(id_vars='Model', var_name='Region', value_name='R-squared')
mae_melted = mae_df.melt(id_vars='Model', var_name='Region', value_name='MAE')
mse_melted = mse_df.melt(id_vars='Model', var_name='Region', value_name='MSE')

# Merge all melted dataframes into one
merged_df = r_squared_melted.merge(mae_melted, on=['Model', 'Region']).merge(mse_melted, on=['Model', 'Region'])

# Create a complex visualization
plt.figure(figsize=(14, 8))

# Scatter plot for R-squared vs MAE
plt.subplot(2, 2, 1)
sns.scatterplot(data=merged_df, x='R-squared', y='MAE', hue='Model', style='Model', s=100)
plt.title('R-squared vs MAE')
plt.xlabel('R-squared')
plt.ylabel('Mean Absolute Error (MAE)')
plt.grid()

# Scatter plot for R-squared vs MSE
plt.subplot(2, 2, 2)
sns.scatterplot(data=merged_df, x='R-squared', y='MSE', hue='Model', style='Model', s=100)
plt.title('R-squared vs MSE')
plt.xlabel('R-squared')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid()

# Line plot for all metrics by Model
plt.subplot(2, 1, 2)
sns.lineplot(data=merged_df.melt(id_vars=['Model'], value_vars=['R-squared', 'MAE', 'MSE']),
             x='variable', y='value', hue='Model', marker="o")
plt.title('Performance Metrics Comparison by Model')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(3, 1, figsize=(8, 15))

axs[0].scatter(r_squared_df['Model'], r_squared_df['NR'], color='blue')
axs[0].plot(r_squared_df['Model'], r_squared_df['NR'], color='blue', linestyle='--')
axs[0].set_title('R-squared Values (NR)') # Update title to reflect the specific region
axs[0].set_xlabel('Models')
axs[0].set_ylabel('R-squared')

# Plot Mean Squared Error (MSE)
# Similarly, for MSE, we need to specify the column name, e.g., 'NR'.
axs[1].scatter(mse_df['Model'], mse_df['NR'], color='orange')
axs[1].plot(mse_df['Model'], mse_df['NR'], color='orange', linestyle='--')
axs[1].set_title('Mean Squared Error (MSE) (NR)') # Update title
axs[1].set_xlabel('Models')
axs[1].set_ylabel('MSE')

# Plot Mean Absolute Error (MAE)
# And for MAE, specify the column name, e.g., 'NR'.
axs[2].scatter(mae_df['Model'], mae_df['NR'], color='green')
axs[2].plot(mae_df['Model'], mae_df['NR'], color='green', linestyle='--')
axs[2].set_title('Mean Absolute Error (MAE) (NR)') # Update title
axs[2].set_xlabel('Models')
axs[2].set_ylabel('MAE')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Plot R-squared Values
for column in r_squared_df.columns[1:]:
    plt.plot(r_squared_df['Model'], r_squared_df[column], marker='o', label=f'R-squared {column}')

# Plot Mean Absolute Error (MAE)
for column in mae_df.columns[1:]:
    plt.plot(mae_df['Model'], mae_df[column], marker='x', linestyle='--', label=f'MAE {column}')

# Plot Mean Squared Error (MSE)
for column in mse_df.columns[1:]:
    plt.plot(mse_df['Model'], mse_df[column], marker='s', linestyle=':', label=f'MSE {column}')

plt.title('Performance Matrix')
plt.xlabel('Models')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```
